```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

df = pd.read_csv('ObesityDataSet_no_duplicates.csv')

categorical_cols = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC',
                    'SMOKE', 'SCC', 'CALC', 'MTRANS', 'NObeyesdad']

le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

correlation_matrix = df.corr(numeric_only=True)
correlation_abs = correlation_matrix.abs()

for i in correlation_abs.columns:
    correlation_abs.loc[i, i] = 0
max_corr_pair = correlation_abs.unstack().sort_values(ascending=False).idxmax()
max_corr_value = correlation_abs.loc[max_corr_pair]

correlations_52 = {
    "SCC_Weight": df["SCC"].corr(df["Weight"]),
    "FAVC_Family": df["FAVC"].corr(df["family_history_with_overweight"]),
    "CH2O_FCVC": df["CH2O"].corr(df["FCVC"]),
    "FAF_Weight": df["FAF"].corr(df["Weight"])
}

spearman_corr, _ = spearmanr(df["Height"], df["Weight"])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def count_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] < lower) | (df[column] > upper)].shape[0]

outliers_height = count_outliers("Height")
outliers_weight = count_outliers("Weight")

results_df = pd.DataFrame({
    "Outliers in Height": [outliers_height],
    "Outliers in Weight": [outliers_weight],
})

print("Outliers:")
display(results_df)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df["Height"])
plt.title("Boxplot of Height")

plt.subplot(1, 2, 2)
sns.boxplot(y=df["Weight"])
plt.title("Boxplot of Weight")

plt.tight_layout()
plt.show()
```

```python
X = df[["Height"]]
y = df["Weight"]

model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

slope, intercept
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

models = {
    "Linear Regression": LinearRegression(),
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    results[name] = {
        "Mean Squared Error": mean_squared_error(y_test, y_pred),
        "R-squared": r2_score(y_test, y_pred)
    }
```

```python

results_df = pd.DataFrame.from_dict(results, orient='index')
print("Model Evaluation Results:")
display(results_df)
```
