```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
```

```python
usage_df = pd.read_csv("wwwusage.csv")
usage_df.rename(columns={'x': 'usage'}, inplace=True)
```

```python
plt.figure(figsize=(9, 4))
plt.plot(usage_df['usage'], linewidth=2, label='Observed Usage')
plt.title("Website Usage Time Series")
plt.xlabel("Time Index")
plt.ylabel("Usage Count")
plt.legend()
plt.grid(True)
plt.show()
```

```python
adf_output = adfuller(usage_df['usage'])
stationary = adf_output[1] <= 0.05
d_param = 0 if stationary else 1
```

```python
order = (1, d_param, 0)
model_instance = ARIMA(usage_df['usage'], order=order)
model_fit = model_instance.fit()
```

```python
next_values = model_fit.forecast(steps=10)
```

```python
print("Forecast for the next 10 time steps:")
for i, value in enumerate(next_values, start=1):
    print(f"Step {i}: {value:.2f}")
```
